# 1. User Login Check

```python
username = "admin"
password = "1234"
# take input from user
user= input("Enter Username: ")
passw= input("Enter Password: ")
# core logic: checks user input with actual username and password
if user == username and passw== password:
  print("Login Successful")
else:
  print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
# students who passed
pass_count=0
# students who failed
fail_count=0
# core logic: iterate through the list and increase the pass_count if marks > 50 otherwise fail_count if marks <50
for i in marks:
  if i >= 50:
    pass_count+=1
  else:
    fail_count+=1
print("Number of Students Passed: ",pass_count,"\n", "Number of Students Failed: ", fail_count)
```

# 3. Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
j=0
# core logic: .strip is inbuilt function that removes spaces from the start and end, .lower is also a inbuilt function that converts uppercase string to lowercase
for i in names:
        names[j]= i.strip().lower()
        j+=1
print(names)
```

# 4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
# core logic: iterate through list and then find the length of string using an inbuilt function len()
for i in messages:
  print("String = ",i, ": length = ", len(i),'\n')
  # if length of string is more than 10 flag it
  if len(i)>10:
    print("Flag: Length of string is more than 10 \n")
```
