# Problem 1: Unique Words in a Sentence

```python
Sentence=input("Enter a sentence: ")
# Core Logic: .split() is an inbuilt function which spits string into list, set() converts list into a unique set no repeating words
unique_wrd = set(Sentence.split())
print("Unique words count: ",len(unique_wrd),"\n", "Unique words: ",unique_wrd)
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
max_ = numbers[0]
min_ = numbers[0]
# Cor Logic: Compare every element of list store in max_ if greater than max_, or less than min_
for i in numbers:
  if i >max_:
    max_=i
  if i<min_:
    min_=i
print("List: ", numbers)
print("Max: ", max_)
print("Min: ",min_)
```

# Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count=0
# Core Logic: loop through each element in list if greater than 1000 increment the count by 1
for i in prices:
  if i > 1000:
    count+=1
print("Products above 1000: ", count)
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
count=0
# Core Logic: Loop through list and compare the corrent element if it is P increase count
for i in attendance:
  if i == "P":
    count+=1
print("Attendance Percentage: ",(count/len(attendance))*100)
```

# Problem 7: Count Character Frequency

```python
string = "pypthon"
count={}
# Core Logic: loop through string every character and check if the current character is already in count: if not set it to 1, if present increase by 1
for i in string:
  if i in count:
    count[i]+=1
  else:
    count[i]=1
print(count)
```

# Problem 9: Check if a Key Exists in a Dictionary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
target = input("Enter Employee Name : ")
# Core Logic: use "in" keyword in python which check if our value or key is in dictionary
if target in employees:
  print("Employee exists")
else:
  print("Employee does NOT exist")
```
